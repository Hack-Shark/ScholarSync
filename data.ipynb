```python
import pandas as pd

column_names = ['title', 'publication_title', 'item_doi', 'authors', 'publication_year', 'url', 'keywords', 'language']

df = pd.read_csv('springer_data.csv',encoding="utf-8",header=None, names=column_names)

# Display the DataFrame
df.head()
```

```python
import re
def remove_non_alphanumeric(row,index):
    if not isinstance(row[index], str):
        return '' 
    pattern = r"[^a-zA-Z0-9 ]"
    return re.sub(pattern, "", row[index]).strip()

# Apply the function to relevant columns
df["title"] = df.apply(remove_non_alphanumeric,index=0,axis=1)
df["publication_title"] = df.apply(remove_non_alphanumeric,index=1,axis=1)
df["authors"] = df.apply(remove_non_alphanumeric,index=3,axis=1)
df["keywords"]=df.apply(remove_non_alphanumeric,index=6,axis=1)
df.head()
```

```python
def extract_numeric(row,index):
    if isinstance(row[index],int):
        return row[index]
    if isinstance(row[index],str):
        pattern = r"[^0-9]"
        return re.sub(pattern, "", row[index]).strip()
    else:
        return ""

df["publication_year"]=df.apply(extract_numeric,index=4,axis=1)
df.head()
```

```python
df.to_csv('output.csv',encoding="utf-8",header=False, index=False)
```
