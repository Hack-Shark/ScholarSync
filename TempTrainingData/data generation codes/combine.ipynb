```python
import pandas as pd 
import numpy

alp='abcdefghijlkmopqrstuvwz'

headers = ['journal_title', 'keyowrds', 'abstract'] 

combined_df = pd.read_csv('n.csv', sep=';',header=None, names=headers)

for x in alp:
   df=pd.read_csv(f'{x}.csv',sep=';',header=None, names=headers)
   combined_df=combined_df.append(df)
    
df1 = pd.read_csv("a_to_z_csvdata.csv")

combined_df.rename(columns = {'journal_title' : '\"Publication Title\"'}, inplace = True)
df1.rename(columns = {' \"Publication Title\"' : '\"Publication Title\"'}, inplace = True)

combined_df.drop(combined_df.loc[combined_df['\"Publication Title\"']=='Critical Care'].index, inplace=True)
combined_df.drop(combined_df.loc[combined_df['\"Publication Title\"']=='Journal of Biomedical Science'].index, inplace=True)
combined_df.drop(combined_df.loc[combined_df['\"Publication Title\"']=='La Rivista del Nuovo Cimento'].index, inplace=True)
combined_df.drop(combined_df.loc[combined_df['\"Publication Title\"']=='Molecules Online'].index, inplace=True)

main = df1.merge(combined_df[['\"Publication Title\"', 'keyowrds']], on = '\"Publication Title\"', how = 'left')

main.rename(columns = {'keyowrds' : 'keywords'}, inplace = True)

main['keywords'] = main['keywords'].replace(numpy.nan, 'NULL')


    
main
```

```python
main.rename(columns = {'"Publication Title"':'Publication Title'}, inplace = True)
main.head()
```

### Language Detection

```python
import langid

# Function to detect language using langid
def detect_language(text):
    lang, _ = langid.classify(text)
    return lang

# Apply language detection function to the Text column
main['Language'] = main['Item Title'].apply(detect_language)
main.head()
```

```python
main=main[main['Language']=='en']
main=main[main['URL']!='Article']
main.describe()
```

```python
fields_to_check = ['Authors']
main = main.dropna(subset=fields_to_check)
null_counts = main.isnull().sum()
print(null_counts)
```

```python
main.to_csv('springer_data.csv', index=False)
```
