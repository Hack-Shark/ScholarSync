```python
import pandas as pd

main=pd.read_csv('springer_data.csv')
main=main[main['Language']=='en']
main=main.drop(['Language'],axis=1)
main.head()
```

```python
main.isnull().sum()
```

```python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from nltk.corpus import stopwords
import unicodedata

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def remove_accents(text):
    text = unicodedata.normalize('NFKD', text).encode('ASCII', 'ignore').decode('utf-8')
    return text

def get_clean_text(row, index):
    if not isinstance(row[index], str):
        return ''
    clean_text = ''
    words = word_tokenize(row[index].lower())
    for word in words:
        # Remove commas
        word = word.replace(',', ' ')
        # Strip accents
        word = remove_accents(word)
        if re.match(r'^[a-zA-Z]+$', word) and len(word) > 1 and word[1] != '.':
            clean_text += ' ' + word
    return clean_text
```

```python
main.columns
```

```python
main['Item Title']=main.apply(get_clean_text,index='Item Title',axis=1)
main['Publication Title']=main.apply(get_clean_text,index='Publication Title',axis=1)
main['Authors']=main.apply(get_clean_text,index='Authors',axis=1)
main
```

```python
main.isnull().sum()
```

```python
null_df = main[(main['Item Title'] == '') | (main['Item DOI'] == '') | (main['Authors'] == '') | (main['Publication Year'] == '') | (main['URL'] == '')]
null_df
```

```python
main = main[~((main['Item Title'] == '') | (main['Publication Title'] == '')| (main['Item DOI'] == '') | (main['Authors'] == '') | (main['Publication Year'] == '') | (main['URL'] == ''))]
main.shape
```

```python
main['Keywords'] = main['Keywords'].fillna('NULL')
main= main.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
new_headers=['item_title','publication_title','item_doi','authors','publication_year','url','keywords']
main.to_csv('cleaned_text.csv',index=False,header=new_headers)
```
